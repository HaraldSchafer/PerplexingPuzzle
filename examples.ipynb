### We can use randomly shuffle the pieces and rotate them to find a brute force solution

```python
from random import shuffle
while True:
  pieces = list(example_grid.flatten())
  shuffle(pieces)
  for i in range(4):
    for k in range(random.randint(0,4)):
      pieces[i].rotate()
  random_grid = np.array(pieces).reshape((2,2))
  if check_grid(random_grid):
    break
print('FOUND GOOD GRID')
imshow(draw_grid(random_grid));
```
